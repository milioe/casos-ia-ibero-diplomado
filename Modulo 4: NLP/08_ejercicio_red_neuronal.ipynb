```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

RANDOM_STATE = 7
COLORES = ["#DC2626", "#2563EB", "#16A34A"]

np.random.seed(RANDOM_STATE)
plt.rcParams.update({"figure.dpi": 100, "axes.grid": True, "grid.alpha": 0.3})

# --- Datos: tres blobs (un poco más dispersos que en el intro: std 1.05 vs 0.85) ---
X, y = make_blobs(
    n_samples=550,
    centers=3,
    cluster_std=3.3,
    random_state=RANDOM_STATE,
)
X = X.astype(np.float32)
y = y.astype(np.int32)

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y,
)

fig, ax = plt.subplots(figsize=(6, 5))
for k, color in zip([0, 1, 2], COLORES):
    m = y == k
    ax.scatter(X[m, 0], X[m, 1], c=color, s=24, alpha=0.7, label=f"Clase {k}")
ax.set_xlabel("x₁")
ax.set_ylabel("x₂")
ax.set_title("Tres grupos — datos del ejercicio")
ax.legend()
plt.tight_layout()
plt.show()
print("Train:", X_train.shape, "| Val:", X_val.shape)
```

```python
def plot_frontera(modelo, X_pts, y_pts, titulo="Frontera de decisión"):
    """Fondo = clase predicha por la red; puntos = datos reales."""
    margen = 0.8
    x_min, x_max = X_pts[:, 0].min() - margen, X_pts[:, 0].max() + margen
    y_min, y_max = X_pts[:, 1].min() - margen, X_pts[:, 1].max() + margen
    paso = 0.05
    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, paso),
        np.arange(y_min, y_max, paso),
    )
    grid = np.c_[xx.ravel(), yy.ravel()].astype(np.float32)
    probs = modelo.predict(grid, verbose=0)
    Z = np.argmax(probs, axis=1).reshape(xx.shape)

    fig, ax = plt.subplots(figsize=(7, 5.5))
    ax.contourf(xx, yy, Z, alpha=0.28, levels=np.arange(-0.5, 3.5, 1), cmap="RdYlBu")
    for k, color in zip([0, 1, 2], COLORES):
        m = y_pts == k
        ax.scatter(
            X_pts[m, 0], X_pts[m, 1], c=color, s=22,
            edgecolors="white", linewidths=0.3, label=f"Clase {k}",
        )
    ax.set_xlabel("x₁")
    ax.set_ylabel("x₂")
    ax.set_title(titulo)
    ax.legend()
    plt.tight_layout()
    plt.show()
```
