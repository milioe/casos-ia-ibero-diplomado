## Verbosity

# Verbosity (Verbosidad)

## ¿Qué es?

Verbosity controla la **cantidad de texto** que genera el modelo en sus respuestas.

## Tres niveles disponibles

### Alta (High)
- Respuestas largas y detalladas
- Código con muchos comentarios y explicaciones
- Útil cuando necesitas entender algo a profundidad

### Media (Medium)
- Balance entre detalle y brevedad
- Era el nivel predeterminado en modelos anteriores
- Adecuado para uso general

### Baja (Low)
- Respuestas cortas y directas
- Código minimalista sin comentarios extras
- Útil cuando solo necesitas la solución rápida

## ¿Cuándo usar cada nivel?

**Alta:** Documentación técnica, refactorización de código, aprendizaje

**Media:** Tareas generales, desarrollo estándar

**Baja:** Queries SQL, snippets simples, respuestas rápidas

## Ventaja y desventaja

- **Menos verbosity** = Respuestas más rápidas (menos tokens)
- **Más verbosity** = Respuestas más lentas pero más completas

## Video

```python
import cv2
import base64

def video_to_frames(video_path, interval_seconds=5):
    """Extrae frames de un video cada N segundos y devuelve sus encodings base64"""
    # Crear directorio temporal si no existe
    temp_dir = "temp_frames"
    os.makedirs(temp_dir, exist_ok=True)
    
    # Abrir el video
    video = cv2.VideoCapture(video_path)
    fps = video.get(cv2.CAP_PROP_FPS)
    frame_interval = int(fps * interval_seconds)
    
    frame_count = 0
    base64_images = []
    
    # Extraer frames
    while True:
        success, frame = video.read()
        if not success:
            break
            
        if frame_count % frame_interval == 0:
            # Guardar frame
            frame_path = f"{temp_dir}/frame_{frame_count}.jpg"
            cv2.imwrite(frame_path, frame)
            
            # Convertir a base64
            with open(frame_path, "rb") as img_file:
                base64_img = base64.b64encode(img_file.read()).decode("utf-8")
                base64_images.append(base64_img)
            
            # Eliminar archivo temporal
            os.remove(frame_path)
            
        frame_count += 1
    
    # Limpiar
    video.release()
    os.rmdir(temp_dir)
    
    return base64_images


base64_frames = video_to_frames(video_path="assets/supermarket.mp4", 
                                interval_seconds=3)

# Preparar contenido para OpenAI
content = [{"type": "input_text", "text": """
            A continuación se te proporcionarán imagenes de supermercado. Tu objetivo es extraer los productos con sus precios con el siguiente formato:
            
            FORMATO:
            Producto: 
            Precio:
            ----
            Producto: 
            Precio:
            ----
            Producto: 
            Precio:
            ----
            FIN DE FORMATO

            """}]

# Añadir cada imagen
for img in base64_frames:
    content.append({
        "type": "input_image",
        "image_url": f"data:image/jpeg;base64,{img}"
    })

# Añadir las reglas - CORREGIDO
content.append({
    "type": "input_text",
    "text": """
    REGLAS:
    - NO inventes información.
    - NO repitas productos, si un mismo producto aparece en dos imagenes no los pongas dos veces.
    - Si el precio de alguno de los productos no es visible, revisa si en la imagen anterior o posterior hay un precio relacionado a ese mismo producto, sino en Precio: escribe "No visible"
    """
})

# Llamar a la API
client = OpenAI(api_key=OPENAI_API_KEY,
                organization=OPENAI_ORG_ID)

response = client.responses.create(
    model="gpt-5",
    input=[{"role": "user", "content": content}],
    max_output_tokens=1_000,
)

print(response.output_text)
```

```python
import requests

def get_weather(latitude, longitude):
    response = requests.get(f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current=temperature_2m,wind_speed_10m&hourly=temperature_2m,relative_humidity_2m,wind_speed_10m")
    data = response.json()
    return data['current']['temperature_2m']


print(get_weather(19.291251, -99.182663))
```

```python
from openai import OpenAI
import json

client = OpenAI(
    api_key=OPENAI_API_KEY,
    organization=OPENAI_ORG_ID
)

tools = [{
    "type": "function",
    "name": "get_weather",
    "description": "Get current temperature for provided coordinates in celsius.",
    "parameters": {
        "type": "object",
        "properties": {
            "latitude": {"type": "number"},
            "longitude": {"type": "number"}
        },
        "required": ["latitude", "longitude"],
        "additionalProperties": False
    },
    "strict": True
}]

input_messages = [{"role": "user", "content": "Cuál es el clima en la CDMX, Tlalpan"}]

response = client.responses.create(
    model="gpt-5",
    input=input_messages,
    tools=tools,
)

print(response.output)
```

```python
import json

tool_call = response.output[1]
args = json.loads(tool_call.arguments)

result = get_weather(args["latitude"], args["longitude"])
print(f"Temperatura en las coordenadas ({args['latitude']}, {args['longitude']}): {result}°C")
```

```python
input_messages.append(tool_call) 

input_messages.append({                            
    "type": "function_call_output",
    "call_id": tool_call.call_id,
    "output": str(result)
})

response_2 = client.responses.create(
    model="gpt-4.1",
    input=input_messages,
    tools=tools,
)
print(response_2.output_text)
```

## Campos a extraer del Currículum

Basándome en el currículum de ejemplo, los campos que deberíamos extraer son:

### Información Personal
- `nombre_completo` (string): Nombre completo del candidato
- `correo` (string): Correo electrónico
- `telefono` (string): Número de teléfono
- `linkedin` (string): URL del perfil de LinkedIn
- `ubicacion` (string): Ciudad/ubicación actual

### Perfil Profesional
- `titulo_profesional` (string): Título o rol principal (ej: "Ingeniero de Datos")
- `anios_experiencia` (integer): Años totales de experiencia
- `resumen_profesional` (string): Breve descripción del perfil

### Experiencia Laboral (array de objetos)
- `puesto` (string): Título del puesto
- `empresa` (string): Nombre de la empresa
- `fecha_inicio` (string): Fecha de inicio
- `fecha_fin` (string): Fecha de fin (o "Presente")
- `responsabilidades` (array of strings): Lista de logros/responsabilidades principales

### Educación (array de objetos)
- `grado` (string): Tipo de grado (Licenciatura, Maestría, etc.)
- `campo_estudio` (string): Área de estudio
- `institucion` (string): Nombre de la institución
- `anio_inicio` (integer): Año de inicio
- `anio_fin` (integer): Año de finalización
- `promedio` (float, opcional): Promedio o GPA

### Habilidades Técnicas (objeto con arrays)
- `lenguajes_programacion` (array of strings)
- `herramientas_big_data` (array of strings)
- `plataformas_cloud` (array of strings)
- `bases_datos` (array of strings)
- `herramientas_devops` (array of strings)
- `herramientas_visualizacion` (array of strings)

### Certificaciones (array de objetos)
- `nombre_certificacion` (string)
- `anio_obtencion` (integer)

### Idiomas (array de objetos)
- `idioma` (string)
- `nivel` (string): Nativo, Avanzado, Intermedio, Básico
